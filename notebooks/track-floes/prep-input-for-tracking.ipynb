### 1. Prepare the working environment

```julia
HOME = "../.." # path to the root of the project two levels up

# Activate the environment
using Pkg
Pkg.activate(HOME)
Pkg.resolve()
Pkg.precompile()
```

```julia
# using IceFloeTracker
using IceFloeTracker: addfloemasks!, imshow, label_components, load, regionprops_table
```

### 2. Load data

```julia
begin # Load data
    image_directory = joinpath(HOME, "explorations", "tracker-imgs") # update to image directory
    img_paths = [path for path in readdir(image_directory, join=true) if endswith(path, ".png")] # edit filter as needed
    images = load.(img_paths)
    labels = label_components.(images)
    props = regionprops_table.(labels)
end

greaterthan0(x) = x .> 0

# Add floe masks
labels_binary = greaterthan0.(labels)
addfloemasks!(props, labels_binary)
```

### 3. View floe masks

```julia
imshow.(props[1].mask)
```
